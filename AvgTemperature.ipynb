```python
data=data.dropna()
```

```python
data['Year'] = data['dt'].str.split('-')
data.drop('dt', axis=1, inplace=True)
data['Year'] = data['Year'].str[0]
data['Year'] = data.Year.astype('int')
```

```python
usa = data.loc[data['Country'] == "United States"]
usa
```

```python
df=pd.DataFrame(data)
usa = df[df['Country'] == "United States"]
usa= usa[usa.Year >= 1953]
usa= usa[usa.Year < 2013]
usa
```

```python
usa = usa.dropna()
df2=pd.DataFrame(usa)
df2
```

```python
x=df2.State.unique()
x = pd.DataFrame(x)
x
```

```python
df2.index.size 
n = len(df2)
n
```

```python
n = len(df2)
avg_temp = []
start = 0
end = start + 12
def avgtemp():
    start = 0
    end = start + 12
    while(start<n):
        result = df2.iloc[range(start,end),[0]]
        temp = round(result['AverageTemperature'].mean(),3)
        avg_temp.append(temp)
        start = start+12
        end = start + 12
avgtemp()
print(avg_temp)
```

```python
avg = avg_temp
avg = pd.DataFrame(avg)
avg
```
